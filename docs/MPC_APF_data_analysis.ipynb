# Data Analysis MPC vs APF

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind

# URLs for the datasets
url_apf = 'https://raw.github.com/davidenascivera/II2202_Project/main/data/apf_summary_metrics.csv'
url_mpc = 'https://raw.github.com/davidenascivera/II2202_Project/main/data/mpc_summary_metrics.csv'

# Define the values to be considered as NaN
na_values = [' None', 'None', '']

# Load the datasets into pandas DataFrames
df_apf = pd.read_csv(url_apf, na_values=na_values, skipinitialspace=True)
df_mpc = pd.read_csv(url_mpc, na_values=na_values, skipinitialspace=True)


# Add a column to indicate the method used
df_apf['Method'] = 'APF'
df_mpc['Method'] = 'MPC'
```

```python
# Identify missing values in df_mpc
print("Missing values in df_mpc:")
print(df_mpc.isnull().sum())

# Drop rows with NaN values in essential columns
essential_columns = ['Total Path Length', 'Final Distance to Target', 'Minimum Distance to Obstacles',
                     'Total Control Effort', 'Average Computation Time per Iteration',
                     'Total Number of Iterations', 'L2 Norm of Control Input Changes', 'Success']

df_mpc_clean = df_mpc.dropna(subset=essential_columns).reset_index(drop=True)

# For consistency, check df_apf for missing values
print("\nMissing values in df_apf:")
print(df_apf.isnull().sum())

# Assuming df_apf has no missing values, but if it does:
df_apf_clean = df_apf.dropna(subset=essential_columns).reset_index(drop=True)
```

```python
# Combine the cleaned DataFrames
df_combined = pd.concat([df_apf_clean, df_mpc_clean], ignore_index=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Filter out rows where 'Total Path Length' is missing or zero
df_filtered = df_combined[df_combined['Total Path Length'].notna() & (df_combined['Total Path Length'] != 0)]

# Further filter to exclude APF rows with Success=False
df_filtered = df_filtered[~((df_filtered['Method'] == 'APF') & (df_filtered['Success'] == False))]

# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Use the same custom palette based on 'coolwarm' for consistency
coolwarm_colors = sns.color_palette("coolwarm", 2)
custom_palette = {"MPC": coolwarm_colors[0], "APF": coolwarm_colors[1]}

# Create a barplot
plt.figure(figsize=(12, 6))
sns.barplot(
    data=df_filtered,
    x='Map number',
    y='Total Path Length',
    hue='Method',
    ci=None,  # Removes error bars
    palette=custom_palette  # Apply the custom color palette
)

# Customize the plot
plt.title('Total Path Length by Map and Method')
plt.xlabel('Map')
plt.ylabel('Total Path Length')
plt.xticks(rotation=45)  # Rotate x-axis labels if needed
plt.legend(title='Method')
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Group by the 'Method' column and compute mean and standard deviation
summary_stats = df_combined.groupby('Method').agg(['mean', 'std'])

# Display the summary statistics
print(summary_stats)
```

```python
# Calculate success rate
success_rates = df_combined.groupby('Method')['Success'].mean()

# Display the success rates
print("\nSuccess Rates:")
print(success_rates)
```

```python
# Metrics to compare
metrics = ['Total Path Length', 'Final Distance to Target', 'Minimum Distance to Obstacles',
           'Total Control Effort', 'Average Computation Time per Iteration',
           'Total Number of Iterations', 'L2 Norm of Control Input Changes']

# Compute mean and standard deviation for each metric
for metric in metrics:
    apf_mean = df_apf_clean[metric].mean()
    apf_std = df_apf_clean[metric].std()
    mpc_mean = df_mpc_clean[metric].mean()
    mpc_std = df_mpc_clean[metric].std()
    print(f"\nMetric: {metric}")
    print(f"APF - Mean: {apf_mean:.2f}, Std: {apf_std:.2f}")
    print(f"MPC - Mean: {mpc_mean:.2f}, Std: {mpc_std:.2f}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style
sns.set(style="whitegrid")

# Extract colors from the 'coolwarm' palette and assign them to MPC and APF
coolwarm_colors = sns.color_palette("coolwarm", 2)
custom_palette = {"MPC": coolwarm_colors[0], "APF": coolwarm_colors[1]}

# Create box plots for each metric with the consistent palette
for metric in metrics:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Method', y=metric, data=df_combined, palette=custom_palette)
    plt.title(f'Comparison of {metric} between APF and MPC')
    plt.show()
```

```python
# Perform t-tests for each metric
for metric in metrics:
    apf_values = df_apf_clean[metric]
    mpc_values = df_mpc_clean[metric]
    t_stat, p_value = ttest_ind(apf_values, mpc_values, equal_var=False, nan_policy='omit')
    print(f"\nMetric: {metric}")
    print(f"T-statistic: {t_stat:.2f}, P-value: {p_value:.4f}")
    if p_value < 0.05:
        print("Result: Statistically significant difference")
    else:
        print("Result: No statistically significant difference")
```
